### Preprocesando los datos

```python
hogares_df2 = hogares_df #pero antes de pasar las variables a números, hacemos una copia del dataset por seguridad
hogares_df2.columns.values
```

```python
#Vamos a crear un dataframe con las columnas que nos interesan
hogares_var_interes = hogares_df2[["PCH1","PCH4","PCH5","PCH6","PCH7","PCH8","PCH9_A","PCH9_B","PCH9_C","PCH9_D",
                                "PCH9_E","PCH9_F","PCH9_G","PCH9_H","PCH9_I","PCH9_K","PCH9_L","PCH9_M","PCH10",
                               "PCH11","PCH12","PCH13","PCH14"]]
hogares_var_interes
```

```python
hogares_var_interes.dtypes
#Vemos que la variable PCH6 está almacenada como un objeto, por lo que la debemos pasar a número, como fue mencionado antes
```

```python
hogares_var_interes = hogares_var_interes.apply(pd.to_numeric,errors='coerce')
#pasamos las variables a números, esta función automáticamente asigna int o float
#Observamos que PCH6 ahora es un número float
hogares_var_interes.dtypes
```

```python
hogares_var_interes.isna().sum()
#Ahora vemos que pasar a número nos generó el problema de que ahora hay 156,927 valores nulos en la variable PCH6
#Vamos a proceder a sustituir los NA por ceros (0)
```

```python
#La variable PCH6 sí la debemos tratar porque es muy importante para nuestro modelo, ya que es la que indica si el servicio
#sanitario en los hogares es privado o compartido
hogares_var_interes["PCH6"].value_counts()
```

```python
hogares_var_interes=hogares_var_interes.fillna(0) #sustituimos los NA por ceros (0)
hogares_var_interes["PCH6"].value_counts() #Listo
```

```python
determinantes_x = hogares_var_interes.to_numpy()
print(hogares_var_interes.shape)
print(determinantes_x.shape)
```

```python
inercia = []
for i in range (1,20):
    nivel_socieconomico = KMeans(n_clusters = i, init = "k-means++", max_iter = 300, n_init = 10)
    nivel_socieconomico.fit(determinantes_x)
    inercia.append(nivel_socieconomico.inertia_)
    
plt.figure(figsize=[10,6])
plt.title("Codo")
plt.xlabel("# de clústers")
plt.ylabel("Inercia")
plt.plot(list(range(1,20)), inercia, marker="o")
plt.show()
```
